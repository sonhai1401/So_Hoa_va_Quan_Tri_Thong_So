# Numpy

```python
import numpy as np

# Tạo một mảng 1 chiều
mang_1d = np.array([1, 2, 3, 4, 5, 6, 7, 8])
print(mang_1d)
# Tạo một mảng 2 chiều
mang_2d = np.array([[1, 3, 5, 7],[2, 4 ,6 ,8]])
print(mang_2d)
# Tạo mảng ngẫu nhiên
mang_nD = np.random.rand(4,4)
print(mang_nD)
```

# Các thao tác cơ bản trong numpy

```python
# Tạo một
mang_1d = np.array([1, 2, 3, 4, 5, 6, 7, 8])
print(mang_1d)

# Cộng và nhân mảng với một số
print(mang_1d + 5)
print(mang_1d * 5)

# Tính tổng, trung bình, lớn nhất, nhỏ nhất
print(mang_1d.sum())
print(mang_1d.mean())
print(mang_1d.max())
print(mang_1d.min())

# Lấy phần tử
print(mang_1d[2])
```

# Pandas

# 2.1 Tạo DataFrame

```python
import pandas as pd

# Tạo DataFrame và dictionary

data = {
    'Tên': ['Anh', 'Yêu', 'Em'],
    'Tuổi': [2006, 2009, 2011],
    'Điểm': [3, 7, 10]
}

df = pd.DataFrame(data)
df
```

## 2.2 Thao tác cơ bản

```python
# Lấy ra hàng tuổi
print(df['Tuổi'])

# Lấy ra hàng thứ hai
print(df.iloc[1])

# Lấy ra người có tuổi lớn hơn 18
print(df[df['Tuổi'] < 2010])

# Lấy điểm trung bình của người đó
print(df['Điểm'].mean())
```

# Matplotlib


## 3.1 Biểu đồ đường

```python
import matplotlib.pyplot as plt

x = [5,8,7,4,32,99]
y = [20,87,64,55,30,100]

plt.plot(x,y,label = 'Biểu đồ đường', color='red', marker='o')
plt.title('Biểu đồ đường')
plt.xlabel('Trục X')
plt.ylabel('Trục Y')
plt.legend()
plt.show()
```

## Biểu đồ cột 

```python
Tên = ["Hải", "Việt", "Cường", "Đại", "Nguyên"]
Điểm = [10, 2.5, 1, 6.5, 3]

plt.bar(Tên, Điểm, color='blue')
plt.title("Biểu đồ cột")
plt.xlabel("Tên")
plt.ylabel("Điểm")
plt.show()
```

## 3.3 Biểu đồ phân tán

```python
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.rand(100)*100

plt.scatter(x,y,s=sizes, alpha=0.5, c='green')
plt.title("Biểu đồ phân tán")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.show()
```
